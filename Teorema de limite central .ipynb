```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import uniform
from scipy.stats import expon
```

```python
#inicializamos dataframe 
poblacion=pd.DataFrame()
poblacion['numbers']=expon.rvs(40,size=10000) #rvs random variable sample
```

```python
poblacion['numbers'].hist(bins=1000) 
# podemos aplicar un estadistico de fisher 
```

```python
muestra_promedio=[]
tamano=5000 # aumentamos mas valor mas clara sera la campana gaussiana
for i in range (0,tamano):
    muestra_promedio.append(poblacion.sample(n=100).mean().values[0])
```

```python
fig,ax=plt.subplots()
ax.hist(muestra_promedio,bins=100,alpha=0.5)
ax.set_xlabel('Valor promedio')
ax.set_ylabel('Frecuencia')
ax.grid() # muestra coordenadas
```
