```python
import pandas as pd
import seaborn as sns
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# variables aleatorias 
from scipy.stats import binom
p=0.3 # probabilidad
n=8 #exitos
x = list(range(0,9,1))
y = list([])

for i in x:
    y.append(binom.pmf(i,p=p, n=n)) #pmf probability mas function calcular probabilidad asociada

#->y # probabilidades  de funcion
#tiene las probabilidades en y graficamos  
fig,ax=plt.subplots()
ax.bar(x,y)
ax.set_ylabel('probabilidad discreta') 


np.sum(y)
```

```python
# media y varianza asociada a mi variable X
#promedio ponderado

media = np.average(x, weights = y)

varianza = np.average(((x - media)**2), weights = y )
media,varianza
fig,ax=plt.subplots()
ax.bar(x,y)
ax.set_ylabel('probabilidad discreta') 
ax.axvline(x=media,c='r',linestyle='--', label = 'valor Esperado')
ax.axvline(x=media+3*np.sqrt(varianza),c='r',linestyle='--', label = 'Desviacion Estandar')
ax.legend()
```
