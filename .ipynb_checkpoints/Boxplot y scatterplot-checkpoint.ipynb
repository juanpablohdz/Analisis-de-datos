```python
#diagrama de caja o box plot 
#y=df['bicis-compartidas']
#ax axes y sns seaborn 
import seaborn as sns
fig, ax =plt.subplots()
sns.boxplot(x ='cuartil-ano', y = 'bicis-compartidas' , data=df)
```

```python
#muestra mas concentracion de puntos o datos 
fig, ax =plt.subplots()
ax.scatter(df['viento'],df['bicis-compartidas'],alpha= 0.1)
ax.set_title('distribucion conjunta de viento y bicis compartidas')
ax.set_xlabel('viento')
ax.set_ylabel('bicis-compartidas')
```
